```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
file_path = '/Users/rush/Documents/Data/linkedin_poster_data_analytics_skills.txt'

with open(file_path, 'r') as file:
    text = file.read()

colors = ["#3399FF", "#33CC33", "#FF6666", "#FFCC00", "#9966CC"]
cmap = LinearSegmentedColormap.from_list("custom vibrant", colors, N=30)

wordcloud = WordCloud(background_color='white',
                      width=1584,
                      height=396,
                      colormap=cmap,
                      collocations=False,
                      stopwords=STOPWORDS,
                      max_font_size=150,
                      min_font_size=20).generate(text)

fig, ax = plt.subplots(figsize=(15.84, 3.96))

ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")

image_path = "/Users/rush/Documents/Data/LinkedIn_Cover_WordCloud.png"
fig.savefig(image_path, format='png', dpi=100, bbox_inches='tight', pad_inches=0)

plt.show()
```
